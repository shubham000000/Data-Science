```python
def detect_outlier(df):
    for i in df.describe().columns:
        Q1 = df.describe().at['25%',i]
        Q3 = df.describe().at['75%',i]
        IQR = Q3 - Q1
                
        LTV = Q1 - 1.5*IQR 
        UTV = Q3 + 1.5*IQR
        df[i] = df[i].mask(df[i]<LTV,LTV)
        df[i] = df[i].mask(df[i]>UTV,UTV)
    return df
```
