```python
import matplotlib.pyplot as plt
from sklearn import datasets
!pip install mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import numpy as np
#import seaborn as sns; sns.set()
```

```python
iris = datasets.load_iris()
```

```python
ss = pd.DataFrame(iris['data'])
```

```python
ss.columns = iris['feature_names']
```

```python
ss.head()
```

```python
#ax = sns.scatterplot(, ], data=ss)
x=ss['sepal length (cm)']
y=ss['sepal width (cm)']
plt.scatter(x, y, alpha=0.8)
plt.title('Iris 2d scatter plot ')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
plt.show()
```

```python
ax = plt.axes(projection='3d')

zdata = ss['sepal length (cm)']
xdata = ss['sepal width (cm)']
ydata = ss['petal length (cm)']
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens');
```
